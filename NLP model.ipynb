```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('main_data.csv')
```

```python
data
```

```python
import nltk
#Stemming and lemmatization
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
#         print(y)
        
    return " ".join(y)
```

```python
#apply stem function to all the rows in tags collum
data['tags'] = data['tags'].apply(stem)
```

```python
data['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# get the top 5000 most common word in tags
# remove all the stop words Ex: is, a, to, etc...
cv = CountVectorizer(max_features=5000, stop_words='english')
# cv = CountVectorizer()
```

```python
##implement countVertor
#cv.fit_transform(data['tags']).toarray().shape

##First movie
# cv.fit_transform(data['tags']).toarray()[0]
# cv.fit_transform(data['tags']).toarray()[0][412]


cv.fit_transform(data['tags']).toarray()
```

```python
## print index of '3d' in tags of the first movie
# cv.get_feature_names().index('battle')

## print all top 5000 most common word
cv.get_feature_names()
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#create 5000 dimension of 2D space (using word)
#In each 2D of dimension ( between 2 words ), 
#we have the vector for every movie ( using bags of word: frequency repeating)
#calcualate the similarity of each movie with the total movie in each pair of dimension
#1 is exactly the same
#0 is completely different

# cosine_similarity(vectors)[0]
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
#similarity of the first movie - Avatar (index 0)
similarity[0]
```

```python
#if we only sort using the similarity at each index, we will mess up the index of each movie
# sorted(similarity[0], reverse = True)

#using enumerate to store index number to keep track of the movie
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x : x[1])
```

```python
#get only five movie to recommend
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x : x[1])[1:6]
```

```python
#recommend function
def recommend(movie):
    #find the index of the movie 
    movie_index = data[data['title'] == movie].index[0]
    
    #find distance between the input movie with other movies
    distance = similarity[movie_index]
    
    #find 5 recommended movies list
    movies_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x : x[1])[1:6]
    
    for i in movies_list:
        print(data.iloc[i[0]].title)
#         print(data.iloc[i[0]].movie_id)
```
